```python
import os
import zipfile
import json
import shutil
import albumentations as A
import cv2
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.nn import Parameter
import timm
import random
import math
from collections import Counter
from tqdm import tqdm
import pandas as pd
from sklearn.model_selection import train_test_split
from torch.optim.lr_scheduler import CosineAnnealingLR

# ============================================================
# 1. ЗАПРОС ДОСТУПА К GOOGLE DRIVE
# ============================================================
from google.colab import drive
drive.mount('/content/drive')

# Создаем папку для результатов на Google Drive
SAVE_DIR = "/content/drive/MyDrive/lab5_metric_learning_final"
os.makedirs(SAVE_DIR, exist_ok=True)
os.makedirs(os.path.join(SAVE_DIR, "checkpoints"), exist_ok=True)
os.makedirs(os.path.join(SAVE_DIR, "submissions"), exist_ok=True)
print(f"Результаты будут сохранены в: {SAVE_DIR}")

# ============================================================
# 2. ФИКСАЦИЯ СИДОВ
# ============================================================
SEED = 9999
os.environ["PYTHONHASHSEED"] = str(SEED)

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = True

def seed_worker(worker_id):
    worker_seed = SEED + worker_id
    np.random.seed(worker_seed)
    random.seed(worker_seed)

g = torch.Generator()
g.manual_seed(SEED)

# ============================================================
# 3. НАСТРОЙКА KAGGLE И ЗАГРУЗКА ДАННЫХ
# ============================================================
!pip install -q kaggle

kaggle_creds = {"username": "birvika", "key": "15d2dbff346b377b89dab3f99d4fadd4"}
os.makedirs('/root/.kaggle', exist_ok=True)
with open('/root/.kaggle/kaggle.json', 'w') as f:
    json.dump(kaggle_creds, f)
!chmod 600 ~/.kaggle/kaggle.json

# Скачиваем RP2K датасет
print("\n" + "="*60)
print("СКАЧИВАНИЕ RP2K ДАТАСЕТА ДЛЯ ПРЕДОБУЧЕНИЯ")
print("="*60)

!kaggle datasets download -d khyeh0719/rp2k-dataset -p /content/ --force

rp2k_zip = "/content/rp2k-dataset.zip"
if os.path.exists(rp2k_zip):
    with zipfile.ZipFile(rp2k_zip, 'r') as zip_ref:
        zip_ref.extractall("/content/rp2k")
    os.remove(rp2k_zip)

# Скачиваем основной датасет соревнования
print("\n" + "="*60)
print("СКАЧИВАНИЕ ДАННЫХ СОРЕВНОВАНИЯ")
print("="*60)

competition_name = "dl-lab-5-metric-learning"
!kaggle competitions download -c {competition_name} -p /content/ --force

main_zip = "/content/dl-lab-5-metric-learning.zip"
if os.path.exists(main_zip):
    with zipfile.ZipFile(main_zip, 'r') as zip_ref:
        zip_ref.extractall("/content/")
    os.remove(main_zip)

train_zip = "/content/train.zip"
if os.path.exists(train_zip):
    with zipfile.ZipFile(train_zip, 'r') as zip_ref:
        zip_ref.extractall("/content/")
    os.remove(train_zip)

test_zip = "/content/test_kaggle.zip"
if os.path.exists(test_zip):
    with zipfile.ZipFile(test_zip, 'r') as zip_ref:
        zip_ref.extractall("/content/")
    os.remove(test_zip)

# Определяем пути
TRAIN_PATH = "/content/train/train"
TEST_PATH = "/content/test_kaggle/test_kaggle"
SUBMISSION_PATH = "/content/submission.csv"

print(f"train_path = '{TRAIN_PATH}'")
print(f"test_path = '{TEST_PATH}'")

# ============================================================
# 4. УСТРОЙСТВО
# ============================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"\nDevice: {device}")
if torch.cuda.is_available():
    print(f"GPU: {torch.cuda.get_device_name(0)}")

# ============================================================
# 5. ФУНКЦИИ ДЛЯ ОБРАБОТКИ ИЗОБРАЖЕНИЙ
# ============================================================
def letterbox(image, target_size=224, color=(0, 0, 0)):
    h, w = image.shape[:2]
    scale = target_size / max(h, w)
    new_w, new_h = int(w * scale), int(h * scale)
    resized = cv2.resize(image, (new_w, new_h), interpolation=cv2.INTER_LINEAR)
    padded = np.full((target_size, target_size, 3), color, dtype=np.uint8)
    y_offset, x_offset = (target_size - new_h) // 2, (target_size - new_w) // 2
    padded[y_offset:y_offset+new_h, x_offset:x_offset+new_w] = resized
    return padded

# ============================================================
# 6. ADACOS LOSS
# ============================================================
class AdaCos(nn.Module):
    def __init__(self, num_features, num_classes, m=0.40):
        super().__init__()
        self.num_features = num_features
        self.n_classes = num_classes
        self.s = math.sqrt(2) * math.log(num_classes - 1)
        self.m = m
        self.W = nn.Parameter(torch.FloatTensor(num_classes, num_features))
        nn.init.xavier_uniform_(self.W)

    def forward(self, input, label=None):
        x = F.normalize(input)
        W = F.normalize(self.W)
        logits = F.linear(x, W)
        if label is None:
            return logits

        theta = torch.acos(torch.clamp(logits, -1.0 + 1e-7, 1.0 - 1e-7))
        one_hot = torch.zeros_like(logits)
        one_hot.scatter_(1, label.view(-1, 1).long(), 1)
        with torch.no_grad():
            B_avg = torch.where(one_hot < 1, torch.exp(self.s * logits), torch.zeros_like(logits))
            B_avg = torch.sum(B_avg) / input.size(0)
            theta_med = torch.median(theta[one_hot == 1])
            self.s = torch.log(B_avg) / torch.cos(torch.min(math.pi/4 * torch.ones_like(theta_med), theta_med))
        output = self.s * logits
        return output

# ============================================================
# 7. МОДЕЛЬ
# ============================================================
class MetricModel(nn.Module):
    def __init__(self, backbone_name='convnext_tiny', embedding_size=512, num_classes=None, pretrained_path=None):
        super().__init__()
        self.backbone = timm.create_model(backbone_name, pretrained=False, num_classes=0)
        in_features = self.backbone.num_features
        self.embedding = nn.Linear(in_features, embedding_size)
        self.bn = nn.BatchNorm1d(embedding_size)
        self.adacos = AdaCos(embedding_size, num_classes, m=0.40) if num_classes else None

        if pretrained_path and os.path.exists(pretrained_path):
            print(f"Загрузка предобученных весов из {pretrained_path}")
            state_dict = torch.load(pretrained_path, map_location='cpu')
            state_dict.pop('adacos.W', None)
            self.load_state_dict(state_dict, strict=False)
            print("Предобученные веса загружены")

    def forward(self, x, labels=None):
        features = self.backbone(x)
        embeddings = self.embedding(features)
        embeddings = self.bn(embeddings)
        embeddings_norm = F.normalize(embeddings, p=2, dim=1)
        if labels is not None and self.adacos is not None:
            logits = self.adacos(embeddings_norm, labels)
            return logits, embeddings_norm
        else:
            return embeddings_norm

# ============================================================
# 8. ДАТАСЕТ
# ============================================================
class ProductDataset(Dataset):
    def __init__(self, images_filepaths, labels, transform=None):
        self.images_filepaths = images_filepaths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.images_filepaths)

    def __getitem__(self, idx):
        image_filepath = self.images_filepaths[idx]
        data = np.fromfile(image_filepath, dtype=np.uint8)
        image = cv2.imdecode(data, cv2.IMREAD_COLOR)
        if image is None:
            image = np.zeros((224, 224, 3), dtype=np.uint8)
        else:
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = letterbox(image, target_size=224)
        if self.transform is not None:
            transformed = self.transform(image=image)
            image = transformed['image']
        return image, self.labels[idx]

# ============================================================
# 9. ПРЕДОБУЧЕНИЕ НА RP2K (7 ЭПОХ)
# ============================================================
print("\n" + "="*60)
print("ПРЕДОБУЧЕНИЕ НА RP2K ДАТАСЕТЕ ")
print("="*60)

# Определяем путь к RP2K
RP2K_PATH = "/content/rp2k/rp2k_dataset/all/train"
if not os.path.exists(RP2K_PATH):
    RP2K_PATH = "/content/rp2k/rp2k_dataset/train"
if not os.path.exists(RP2K_PATH):
    RP2K_PATH = "/content/rp2k/train"
if not os.path.exists(RP2K_PATH):
    for root, dirs, files in os.walk("/content/rp2k"):
        if "train" in dirs:
            RP2K_PATH = os.path.join(root, "train")
            break

print(f"RP2K путь: {RP2K_PATH}")

rp2k_classes = [d for d in os.listdir(RP2K_PATH) if os.path.isdir(os.path.join(RP2K_PATH, d))]
rp2k_classes.sort()
rp2k_class_to_idx = {cls: idx for idx, cls in enumerate(rp2k_classes)}

rp2k_paths, rp2k_labels = [], []
for class_name in tqdm(rp2k_classes, desc="Загрузка RP2K классов"):
    class_dir = os.path.join(RP2K_PATH, class_name)
    for img_name in os.listdir(class_dir):
        if img_name.lower().endswith(('.jpg', '.jpeg', '.png')):
            rp2k_paths.append(os.path.join(class_dir, img_name))
            rp2k_labels.append(rp2k_class_to_idx[class_name])

print(f"RP2K: {len(rp2k_paths)} изображений, {len(rp2k_classes)} классов")

# Аугментации для RP2K
rp2k_transform = A.Compose([
    A.Rotate(limit=180, p=0.3),

    A.RandomBrightnessContrast(p=0.2),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    A.ToTensorV2(),
])

rp2k_dataset = ProductDataset(rp2k_paths, rp2k_labels, transform=rp2k_transform)
rp2k_loader = DataLoader(rp2k_dataset, batch_size=128, shuffle=True, num_workers=2, pin_memory=True)

pretrain_model = MetricModel(
    backbone_name='convnext_tiny',
    embedding_size=512,
    num_classes=len(rp2k_classes)
).to(device)

optimizer = torch.optim.AdamW(pretrain_model.parameters(), lr=2e-4, weight_decay=0.05)
scheduler = CosineAnnealingLR(optimizer, T_max=7, eta_min=1e-6)

best_pretrain_loss = float('inf')
PRETRAIN_EPOCHS = 1

for epoch in range(1, PRETRAIN_EPOCHS + 1):
    pretrain_model.train()
    total_loss = 0
    num_batches = 0

    pbar = tqdm(rp2k_loader, desc=f"Pretrain Epoch {epoch}/{PRETRAIN_EPOCHS}")
    for images, labels in pbar:
        images, labels = images.to(device), labels.to(device)
        logits, _ = pretrain_model(images, labels)
        loss = F.cross_entropy(logits, labels)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        total_loss += loss.item()
        num_batches += 1
        avg_loss = total_loss / num_batches
        pbar.set_postfix(loss=f"{avg_loss:.4f}")

    scheduler.step()
    avg_loss = total_loss / len(rp2k_loader)

    if avg_loss < best_pretrain_loss:
        best_pretrain_loss = avg_loss
        torch.save(pretrain_model.state_dict(), os.path.join(SAVE_DIR, "pretrained_rp2k.pth"))
        print(f"  ★ Сохранена лучшая предобученная модель, Loss={avg_loss:.4f}")

PRETRAINED_PATH = os.path.join(SAVE_DIR, "pretrained_rp2k.pth")
print(f"\nПредобучение завершено! Модель сохранена: {PRETRAINED_PATH}")

# ============================================================
# 10. ПОДГОТОВКА ОСНОВНОГО ДАТАСЕТА
# ============================================================
print("\n" + "="*60)
print("ПОДГОТОВКА ОСНОВНОГО ДАТАСЕТА")
print("="*60)

class_folders = [d for d in os.listdir(TRAIN_PATH) if os.path.isdir(os.path.join(TRAIN_PATH, d))]
class_folders.sort()
class_to_idx = {cls: idx for idx, cls in enumerate(class_folders)}
num_classes = len(class_to_idx)

print(f"Классов: {num_classes}")

all_paths, all_labels = [], []
for class_name in tqdm(class_folders, desc="Загрузка классов"):
    class_dir = os.path.join(TRAIN_PATH, class_name)
    for img_name in os.listdir(class_dir):
        if img_name.lower().endswith(('.jpg', '.jpeg', '.png')):
            all_paths.append(os.path.join(class_dir, img_name))
            all_labels.append(class_to_idx[class_name])

print(f"Всего изображений: {len(all_paths)}")

# ============================================================
# 11. РАЗБИЕНИЕ
# ============================================================
train_paths, val_paths, train_labels, val_labels = train_test_split(
    all_paths, all_labels, test_size=0.2, random_state=SEED, stratify=all_labels
)

unique_labels = sorted(set(train_labels))
label_map = {old: new for new, old in enumerate(unique_labels)}
train_labels_mapped = [label_map[l] for l in train_labels]
val_labels_mapped = [label_map[l] for l in val_labels]

print(f"Train: {len(train_paths)}, Val: {len(val_paths)}")

# ============================================================
# 12. ТРАНСФОРМАЦИИ
# ============================================================
train_transform = A.Compose([
    A.Rotate(limit=180, p=0.11),
    A.RandomBrightnessContrast(p=0.11),
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    A.ToTensorV2(),
])

val_transform = A.Compose([
    A.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    A.ToTensorV2(),
])

train_dataset = ProductDataset(train_paths, train_labels_mapped, transform=train_transform)
val_dataset = ProductDataset(val_paths, val_labels_mapped, transform=val_transform)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True, num_workers=2, pin_memory=True, worker_init_fn=seed_worker, generator=g)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False, num_workers=2, pin_memory=True, worker_init_fn=seed_worker, generator=g)

# ============================================================
# 13. ФИНАЛЬНОЕ ОБУЧЕНИЕ С ПРЕДОБУЧЕНИЕМ (12 ЭПОХ)
# ============================================================
print("\n" + "="*60)
print("ФИНАЛЬНОЕ ОБУЧЕНИЕ (с предобучением на RP2K) - 11 ЭПОХ")
print("="*60)

model = MetricModel(
    backbone_name='convnext_tiny',
    embedding_size=512,
    num_classes=len(unique_labels),
    pretrained_path=PRETRAINED_PATH
).to(device)

optimizer = torch.optim.AdamW(model.parameters(), lr=5e-4, weight_decay=0.01)
scheduler = CosineAnnealingLR(optimizer, T_max=12, eta_min=1e-6)

best_loss = float('inf')
patience_counter = 0

for epoch in range(1, 12):
    model.train()
    total_loss = 0
    total_samples = 0
    optimizer.zero_grad()

    pbar = tqdm(train_loader, desc=f"Epoch {epoch}/12")
    for images, labels in pbar:
        images, labels = images.to(device), labels.to(device)
        logits, _ = model(images, labels)
        loss = F.cross_entropy(logits, labels)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        total_loss += loss.item() * len(images)
        total_samples += len(images)
        avg_loss = total_loss / total_samples
        pbar.set_postfix(loss=f"{avg_loss:.4f}")

    scheduler.step()
    avg_loss = total_loss / total_samples

    # Валидация
    model.eval()
    val_loss = 0
    val_correct = 0
    val_total = 0
    with torch.no_grad():
        for images, labels in tqdm(val_loader, desc="Validation"):
            images, labels = images.to(device), labels.to(device)
            logits, _ = model(images, labels)
            loss = F.cross_entropy(logits, labels)
            val_loss += loss.item() * len(images)
            val_correct += (logits.argmax(1) == labels).sum().item()
            val_total += len(images)

    avg_val_loss = val_loss / val_total
    val_acc = val_correct / val_total

    print(f"Epoch {epoch}: Train Loss={avg_loss:.4f}, Val Loss={avg_val_loss:.4f}, Val Acc={val_acc:.4f}")

    # Сохраняем чекпоинт
    checkpoint_path = os.path.join(SAVE_DIR, "checkpoints", f'model_epoch_{epoch}.pth')
    torch.save({
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'loss': avg_loss,
        'val_loss': avg_val_loss,
        'val_acc': val_acc
    }, checkpoint_path)
    print(f"  Чекпоинт сохранен: {checkpoint_path}")

    if avg_val_loss < best_loss:
        best_loss = avg_val_loss
        torch.save(model.state_dict(), os.path.join(SAVE_DIR, "best_model.pth"))
        print(f"  ★ Лучшая модель: Val Loss={best_loss:.4f}")
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= 3:
            print(f"Ранняя остановка на эпохе {epoch}")
            break

print(f"\nОбучение завершено! Лучший Val Loss: {best_loss:.4f}")

# ============================================================
# 14. ИНФЕРЕНС (БЕЗ TTA, БЕЗ АНСАМБЛЕЙ)
# ============================================================
print("\n" + "="*60)
print("ИНФЕРЕНС (один проход по изображению)")
print("="*60)

model.load_state_dict(torch.load(os.path.join(SAVE_DIR, "best_model.pth"), map_location=device))
model.eval()

submission = pd.read_csv(SUBMISSION_PATH)
print(f"Загружено пар: {len(submission)}")

if 'similarity' in submission.columns:
    submission = submission.drop(columns=['similarity'])

all_files = list(set(submission['file_1'].tolist() + submission['file_2'].tolist()))
print(f"Уникальных файлов: {len(all_files)}")

embedding_cache = {}

print("Вычисление эмбеддингов...")
for i in tqdm(range(0, len(all_files), 64)):
    batch_files = all_files[i:i+64]
    batch_images = []
    valid_files = []

    for file_name in batch_files:
        img_path = os.path.join(TEST_PATH, file_name)
        if not os.path.exists(img_path):
            for root, dirs, files in os.walk(TEST_PATH):
                if file_name in files:
                    img_path = os.path.join(root, file_name)
                    break

        image = cv2.imread(img_path)
        if image is None:
            embedding_cache[file_name] = np.zeros(512, dtype=np.float32)
            continue

        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = letterbox(image, target_size=224)
        image = image.astype(np.float32) / 255.0
        mean = np.array([0.485, 0.456, 0.406])
        std = np.array([0.229, 0.224, 0.225])
        image = (image - mean) / std
        tensor = torch.from_numpy(image).permute(2, 0, 1).float()
        batch_images.append(tensor)
        valid_files.append(file_name)

    if batch_images:
        batch_tensor = torch.stack(batch_images).to(device)
        with torch.no_grad():
            embeddings = model(batch_tensor).cpu().numpy()
        for file_name, emb in zip(valid_files, embeddings):
            embedding_cache[file_name] = emb.astype(np.float32)

print("Вычисление similarity...")
similarities = []
for _, row in tqdm(submission.iterrows(), total=len(submission)):
    emb1 = embedding_cache.get(row['file_1'], np.zeros(512, dtype=np.float32))
    emb2 = embedding_cache.get(row['file_2'], np.zeros(512, dtype=np.float32))
    sim = np.dot(emb1, emb2)
    similarities.append(float(np.clip(sim, 0.0, 1.0)))

submission['similarity'] = similarities
output_path = os.path.join(SAVE_DIR, "submissions", "predictions_metric_final.csv")
submission.to_csv(output_path, index=False)

print(f"\nРезультат сохранен: {output_path}")
print(f"Similarity: min={np.min(similarities):.4f}, max={np.max(similarities):.4f}, mean={np.mean(similarities):.4f}")

# ============================================================
# 15. СОХРАНЕНИЕ ИТОГОВОГО АРХИВА
# ============================================================
print("\n" + "="*60)
print("СОЗДАНИЕ АРХИВА РЕЗУЛЬТАТОВ")
print("="*60)

# Копируем финальный сабмишен в корень папки
shutil.copy2(output_path, os.path.join(SAVE_DIR, "predictions_metric_final.csv"))

# Создаем архив
archive_path = shutil.make_archive(os.path.join(SAVE_DIR, "lab5_final_results"), 'zip', SAVE_DIR)
print(f"Архив результатов создан: {archive_path}")

# ============================================================
# 16. ВЫВОД ИТОГОВОЙ ИНФОРМАЦИИ
# ============================================================
print("\n" + "="*60)
print("ИТОГИ РАБОТЫ")
print("="*60)

print(f"\nВсе результаты сохранены в: {SAVE_DIR}")

print("\nСодержимое папки:")
for item in os.listdir(SAVE_DIR):
    item_path = os.path.join(SAVE_DIR, item)
    if os.path.isdir(item_path):
        num_files = len([f for f in os.listdir(item_path) if os.path.isfile(os.path.join(item_path, f))])
        print(f"  📁 {item}/ - {num_files} файлов")
    else:
        size_mb = os.path.getsize(item_path) / (1024 * 1024)
        print(f"  📄 {item} ({size_mb:.2f} MB)")

print("\n" + "="*60)
print("ГОТОВО!")
print("="*60)
```
